```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from skimage.io import imread, imshow
from skimage import io, img_as_float

```

```python
image = imread('Y1.jpg', as_gray = True)
image = img_as_float(image)
imshow(image)
print(np.mean(image))
image.sum(0).sum(0)
```
